```python
import requests
from urllib import request # I prefer how urllib.requests handles downloading files
from bs4 import BeautifulSoup
```

We are trying to fetch 2 types of data - Decadal variation in Population from 1901-2011 for each state/district (PC11_A02), and General Economic Tables showing breakdown of Main Workers, Marginal Workers & Non-Workers by Age & Sex (PC11-B01).

```python
base_url_pop = 'https://censusindia.gov.in/nada/index.php/catalog/central/?page=1&sk=PC11_A02&sort_by=popularity&sort_order=desc&ps=50'
base_url_work = 'https://censusindia.gov.in/nada/index.php/catalog/central/?page=1&sk=PC11_B01&sort_by=relevance&sort_order=desc&ps=50'
```

```python
response_pop = requests.get(base_url_pop)
response_work = requests.get(base_url_work)
print(response_pop.status_code, response_work.status_code)
```

```python
html_pop = response_pop.content
print(html_pop[0:100])
html_work = response_work.content
print(html_work[0:100])
```

```python
soup_pop = BeautifulSoup(html_pop, parser = 'html')
soup_pop
```

```python
soup_work = BeautifulSoup(html_work, parser = 'html')
soup_work
```

```python
with open('DownloadedWebpage/CensusIndiaDigLibPop.html', 'wb') as file:
    file.write(soup_pop.prettify('utf-8'))

with open('DownloadedWebpage/CensusIndiaDigLibWork.html', 'wb') as file:
    file.write(soup_work.prettify('utf-8'))
```

```python
classes_dflex_pop = soup_pop.find_all('a', {'class' : 'd-flex'}, )
classes_dflex_pop
```

```python
classes_dflex_work = soup_work.find_all('a', {'class' : 'd-flex'}, )
classes_dflex_work
```

B-01 (Worker) data looks a bit messier than A-02 (Decadal Variance of Population), since seraching for PC11_B01 also returns results for B01 (ST), B01 (SC) etc. which is not what we are looking for at the moment. So, we need to filter out those results and only look for B-01 links. We also need to be careful here, our B-01 files and filecount should match with the A-02 files we are fetching - otherwise we won't be able to join these together later. We are sorting the B-01 files with Relevance (used Popularity for A-02) to prevent ST files from coming on top. Let's see if we can find all the B-01 files in the top 50 links.

Let's first get the A-02 files - that can then tell us which B-01 files we need.

```python
search_links_pop = [class_dflex['href'] for class_dflex in classes_dflex_pop]
country_state_ut_pop = [class_dflex['title'].split(',')[-1].replace(' ', '') for class_dflex in classes_dflex_pop]
```

```python
file_path_pop = 'PopVarFiles/'
for idx in range(len(search_links_pop)):
    print('Getting the file for {} from {}.'.format(country_state_ut_pop[idx], search_links_pop[idx]))
    response_filepage = requests.get(search_links_pop[idx])
    print('Link Response : ', response_filepage.status_code)
    soup_filepage = BeautifulSoup(response_filepage.content, parser = 'html')
    
    # handling for xls or xlsx file extensions
    file_extension = '.xlsx'
    file_type = soup_filepage.select_one('a[href*=xlsx]')
    if(file_type == None):
        file_extension = '.xls'
        file_link = soup_filepage.select_one('a[href*=xls]')['href']
    else:
        file_link = soup_filepage.select_one('a[href*=xlsx]')['href']
    
    print('File Download Link : ', file_link)
    file_name = file_path_pop + country_state_ut_pop[idx] + file_extension
    request.urlretrieve(file_link, file_name)
    print(file_name, ' saved.')
```

We have the A-02 files now, and a list with names of all the states, union territories etc. we got the files for. Let's see wether the B-01 files match these.

```python
# B-01 title example : <need these> B-01: Main workers, marginal workers, non-workers and those marginal workers,
#  non-workers seeking/available for work classified by age and sex (total), West Bengal - 2011
# <don't need these> B-01 (ST): Main workers, marginal workers, non-workers and those marginal workers,
#  non-workers seeking/available for work classified by age and sex (scheduled tribes), Manipur - 2011

search_links_work = [class_dflex['href'] for class_dflex in classes_dflex_work if class_dflex['title'].split(':')[0].strip() == 'B-01']
country_state_ut_wyear_work = [class_dflex['title'].split(',')[-1] for class_dflex in classes_dflex_work if class_dflex['title'].split(':')[0].strip() == 'B-01']
country_state_ut_work = [name.split('-')[0].replace(' ', '') for name in country_state_ut_wyear_work]
```

Let's now check whether the B-01 lists (work) match the A-02 lists(pop).

```python
assert len(search_links_work) == len(search_links_pop)
set(country_state_ut_work) == set(country_state_ut_pop)
```

```python
file_path_work = 'WorkFiles/'
for idx in range(len(search_links_work)):
    print('Getting the file for {} from {}.'.format(country_state_ut_work[idx], search_links_work[idx]))
    response_filepage = requests.get(search_links_work[idx])
    print('Link Response : ', response_filepage.status_code)
    soup_filepage = BeautifulSoup(response_filepage.content, parser = 'html')
    
    # handling for xls or xlsx file extensions
    file_extension = '.xlsx'
    file_type = soup_filepage.select_one('a[href*=xlsx]')
    if(file_type == None):
        file_extension = '.xls'
        file_link = soup_filepage.select_one('a[href*=xls]')['href']
    else:
        file_link = soup_filepage.select_one('a[href*=xlsx]')['href']
    
    print('File Download Link : ', file_link)
    file_name = file_path_work + country_state_ut_work[idx] + file_extension
    request.urlretrieve(file_link, file_name)
    print(file_name, ' saved.')
```
